```python
import pandas as pd
```

```python
matplotlib_commits = 'matplotlib_commits.csv'
matplotlib_issues = 'matplotlib_issues.csv'
pandas_commits = 'pandas_commits.csv'
pandas_issues = 'pandas_issues.csv'
sympy_commits = 'sympy_commits.csv'
sympy_issues = 'sympy_issues.csv'
scikit_commits = 'scikit-learn_commits.csv'
scikit_issues = 'scikit-learn_issues.csv'
```

### A maioria dos commits s√£o merges

```python
df_sympy_commits = pd.read_csv(sympy_commits, sep=';')
df_sympy_commits.sort_values('handlers', ascending=False)
```

### Issues

```python
df_sympy_issues = pd.read_csv(sympy_issues, sep=';')
df_sympy_issues
```

### Commits que n√£o s√£o merge (improvisado, √© preciso excluir os mergesna coleta)

```python
df_scikit_commits = pd.read_csv(scikit_commits, sep=';')
df_scikit_commits[~df_scikit_commits['message'].str.contains('Merge|merge|conflicts')].sort_values('handlers', ascending=False)
```
